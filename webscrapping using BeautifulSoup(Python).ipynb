```python
import requests
from bs4 import BeautifulSoup
```

```python
url="https://shopify.dev/docs"

# we get html files (Shopify)
r= requests.get(url)
htmlContent=r.content
print(htmlContent)
```

```python
# Parse the HTML 
soup=BeautifulSoup(htmlContent,'html.parser')
print(soup.prettify)  # we can see our html file in manner ways using the prettify
```

```python
# html Tree Traversal
title=soup.title
print(title)
```

```python
title =soup.title
print(type(title))
```

```python
# commonly used type of object
# 1) Tag
# 2) navigable String -this string is diffrent from our python string  but we can use speciall case and objects in navigates string
# 3)beautifulSoup
# 4) comment
title=soup.title
print(type(title))
print(type(soup))
print(type(title.string))
```

```python
# Get the Title of the HTML page
title=soup.title
print(title)
```

```python
# get all the paragraphs from the pages
paras = soup.find_all('p')
print(paras)
```

```python
# get all the anchor Tags from the pages
anchors = soup.find_all('a')
print(anchors)
```

```python
# get thev first element in the html pages
print(soup.find('p'))
```

```python
print(soup.find('p').get('class'))
```

```python
# find all element with class lead
print(soup.find_all("p",class_="lead"))
```

```python
# get text from the soup/tags
print(soup.find('p').get_text())
```

```python
# get full html text into tag
print(soup.get_text())
```

```python
# get all links on the page
anchors = soup.find_all('a')
for link in anchors:
    print(link.get("href"))
```

```python
# check the navbar id
for tag in soup.find_all(attrs={"id": True}):
    print(tag['id'])
```
